```python
import pandas as pd
import numpy as np


dfAll_minus = pd.read_csv('../../Cache/All_minus.csv')
dfAll = pd.read_csv('../../Data/All_non_missing_v2.csv')
```

```python
dfAll.columns
```

```python
raw_predictors = dfAll.columns.tolist()[1:-13]
```

```python
dfBase = dfAll.copy()
criteria = 0.1
groupLists = []
for i in range(len(raw_predictors)-1):
    if raw_predictors[i] in dfBase.columns:
        tmpList = [raw_predictors[i]]
    else:
        continue
    for j in range(i+1,len(raw_predictors)):
        mean_gap = float(abs(dfAll[raw_predictors[i]].mean()-dfAll[raw_predictors[j]].mean()))
        range1 = dfAll[raw_predictors[i]].max() - dfAll[raw_predictors[i]].min()
        range2 = dfAll[raw_predictors[j]].max() - dfAll[raw_predictors[j]].min()
        if mean_gap/range1<=criteria and mean_gap/range2<=criteria:
            tmpList.append(raw_predictors[j])
            
    if len(tmpList)>=3:
        groupLists.append(tmpList)
        for var in tmpList:
            try:
                del dfBase[var]
            except:
                continue
    
    
```

```python
t=0
added = ['ID']
for varList in groupLists:
    varList = sorted(varList)
    t+=1 
    dfAll['group_%d_std'%t] = dfAll[varList].std(axis=1)
    '''dfTmp = dfAll[varList].T
    dfTmp['trend'] = range(1,len(varList)+1)
    dfTrend = dfTmp.corr()
    dfTrend = dfTrend.iloc[0:-1,-1]
    dfAll['group_%d_trend'%t] = dfTrend.values.tolist()'''
    added.append('group_%d_std'%t)
    #added.append('group_%d_trend'%t)
    dfAll['group_%d_range'%t] = dfAll[varList].max(axis=1)-dfAll[varList].min(axis=1)
    dfAll['group_%d_unique'%t] = dfAll[varList].nunique(axis=1)/float(len(varList))
    dfAll['group_%d_median_larger'%t] = (dfAll[varList].T>dfAll[varList].median(axis=1)).sum()/float(len(varList))
    added.append('group_%d_range'%t)
    added.append('group_%d_unique'%t)
    added.append('group_%d_median_larger'%t)  
```

```python
dfAll[added].to_csv('../../Cache/group_var.csv')
```
